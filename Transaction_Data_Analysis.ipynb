```python
Product_Sales = df1.groupby(['PROD_NBR', 'PROD_NAME'])['TOT_SALES'].sum().reset_index()
```

```python
Top_Products = Product_Sales.sort_values(by='TOT_SALES', ascending=False).head(3)
```

```python
Top_Products
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='PROD_NAME', y='TOT_SALES', data=Top_Products, palette='husl')
plt.xticks(rotation=40)
plt.title("Top 3 Most Profitable Products Based On Sales")
plt.xlabel("Product Name")
plt.ylabel("Total Sales")
plt.show()
```

```python
Merged_Data = df1.merge(df2, on='LYLTY_CARD_NBR')
```

```python
print(Merged_Data.head(5))
```

```python
Loyal_Customers = Merged_Data.groupby(['LIFESTAGE', 'PREMIUM_CUSTOMER'])['LYLTY_CARD_NBR'].nunique().reset_index()
```

```python
Loyal_Customers = Loyal_Customers.sort_values(by='LYLTY_CARD_NBR', ascending=False)
```

```python
print(Loyal_Customers.head(5))
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='LIFESTAGE', y='LYLTY_CARD_NBR', hue='PREMIUM_CUSTOMER', data=Loyal_Customers, palette='husl')
plt.xticks(rotation=50)
plt.title("Loyal Customers Segments")
plt.xlabel("Life Stage")
plt.ylabel("Number of Customers")
plt.legend(title="Premium Customer Segments")
plt.show()
```
